```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/laptop_data.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
for columns in df.columns:
  df[columns] = LE.fit_transform(df[columns])
```

```python
df.info()
```

```python
final_res=[]
```

```python
x=df.drop(columns=['Price'], axis=1)
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state=42)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(x_train,y_train)
```

```python
y_pred = LR.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score_LR = r2_score(y_test, y_pred)
print(r2_score_LR)
```

```python
final_res.append(r2_score_LR)
```

```python
from sklearn.tree import DecisionTreeRegressor
DTR = DecisionTreeRegressor()
DTR.fit(x_train,y_train)
```

```python
y_pred2 = DTR.predict(x_test)
```

```python
r2_score_DTR = r2_score(y_test, y_pred2)
print(r2_score_DTR)
```

```python
final_res.append(r2_score_DTR)
```

```python
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor()
RFR.fit(x_train,y_train)
```

```python
y_pred3 = RFR.predict(x_test)
```

```python
r2_score_RFR = r2_score(y_test, y_pred3)
print(r2_score_RFR)
```

```python
final_res.append(r2_score_RFR)
```

```python
final_res
```

```python
from sklearn.svm import SVR
SV = SVR()
SV.fit(x_train,y_train)
```

```python
y_pred4 = SV.predict(x_test)
```

```python
r2_score_SV = r2_score(y_test, y_pred4)
print(r2_score_SV) # thought this is advanced mode, it overthinks/overanalyze/overlearn, resulting in low accuracy.
final_res.append(r2_score_SV)
```

```python
from sklearn.neighbors import KNeighborsRegressor
KNR = KNeighborsRegressor()
KNR.fit(x_train,y_train)
```

```python
y_pred5 = KNR.predict(x_test)
```

```python
r2_score_KNR = r2_score(y_test, y_pred5)
print(r2_score_KNR)
final_res.append(r2_score_KNR)
```

```python
from xgboost import XGBRegressor
XGR = XGBRegressor()
XGR.fit(x_train,y_train)
```

```python
y_pred6 = XGR.predict(x_test)
```

```python
r2_score_XGR = r2_score(y_test, y_pred6)
print(r2_score_XGR)
final_res.append(r2_score_XGR)
```

```python
final = np.array(final_res)
result = final.reshape(-1,1)
columns = ['R2_score']
index=['Linear Regression', 'Decision Tree Regressor', 'Random Forest Regressor', 'Support Vector Regressor', 'KNeighbors Regressor','XGBRegressor']

final_result = pd.DataFrame(result, index=index, columns=columns).sort_values(by='R2_score', ascending=False)
final_result
```
